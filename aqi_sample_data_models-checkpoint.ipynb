```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.metrics import root_mean_squared_error, mean_absolute_error, mean_squared_error, r2_score
from sklearn.pipeline import Pipeline
from sklearn.inspection import permutation_importance
import warnings
warnings.filterwarnings("ignore")
rfr_pipeline = Pipeline([
    ("model", RandomForestRegressor(n_estimators=100, max_depth=None , n_jobs=-1,random_state=100)) #
])
svr_pipeline = Pipeline([
    ("model", SVR(kernel="rbf", C=10, epsilon=0.1, gamma="scale"))
])
```

```python
X = df.drop(["aqi_value", "city_name", "country_name", "aqi_category", 
             "co_aqi_category", "ozone_aqi_category", "no2_aqi_category", "pm2.5_aqi_category"], axis=1) #Features
y = df[["aqi_value"]] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) #80% training, 20% testing
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
X_train
```

```python
y_train
```

```python
#xgb
xgb = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=None, subsample=0.8, 
                   colsample_bytree=0.8,  n_jobs=-1, random_state=100)
xgb.fit(X_train, y_train)
```

```python
#rfr
rfr = rfr_pipeline.fit(X_train, y_train)
rfr
```

```python
#svr
svr = svr_pipeline.fit(X_train, y_train)
svr
```

```python
#Rfr
y_pred_rfr = rfr_pipeline.predict(X_test)
y_pred_svr = svr_pipeline.predict(X_test)
y_pred_xgb = xgb.predict(X_test)
```

```python
#Rfr
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred_rfr))
print("Root Mean Squared Error(RMSE):", root_mean_squared_error(y_test, y_pred_rfr))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred_rfr))
print("R2 Score:", r2_score(y_test, y_pred_rfr))
```

```python
#XGB
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred_xgb))
print("Root Mean Squared Error(RMSE):", root_mean_squared_error(y_test, y_pred_xgb))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred_xgb))
print("R2 Score:", r2_score(y_test, y_pred_xgb))
```

```python
#SVR
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred_svr))
print("Root Mean Squared Error(RMSE):", root_mean_squared_error(y_test, y_pred_svr))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred_svr))
print("R2 Score:", r2_score(y_test, y_pred_svr))
```

## Permutation Importance

```python
perm_importance_rfr = permutation_importance(
    rfr_pipeline,
    X_test,
    y_test,
    n_repeats = 10,
    random_state = 100
)
```

```python
perm_importance_df_rfr = pd.DataFrame({
    "Feature" : X.columns,
    "Importance" : perm_importance.importances_mean
})
```

```python
print(perm_importance_df_rfr.sort_values(by='Importance', ascending=False))
```
